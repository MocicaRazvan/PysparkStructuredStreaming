# Words Producer


```python
KAFKA_TOPIC = "bd-words"
KAFKA_BOOTSTRAP_SERVER = 'bd-kafka:29092'
```

Vom trimite mesaje constant catre Kafka, cu un cuvant random si un timestamp random. De asemnea, vom adauga un header `content-type` pentru a specifica tipul de continut al mesajului si un key random pentru a putea identifica mesajul in Kafka. Am adugat si un callback pentru a putea monitoriza trimiterea mesajelor.
Pentru a simula un delay intre mesaje, vom folosi `time.sleep(random.randint(1, 5))`.


```python
from confluent_kafka import Producer
import json
import random
import uuid
import time

words: list = ["apple", "banana", "cherry", "date", "elderberry", "fig", "grape", "honeydew", "kiwi", "lemon", "mango", "nectarine", "orange", "pear", "quince", "raspberry", "strawberry", "tangerine", "watermelon", "apricot", "blackberry", "cantaloupe", "dragonfruit", "grapefruit", "jackfruit", "kiwifruit", "lime", "mulberry", "olive", "papaya", "quince", "rhubarb", "starfruit", "tamarind", "ugli fruit", "vanilla bean", "yuzu", "zucchini", "almond", "cashew", "date", "eggplant", "fig", "garlic", "hazelnut", "iceberg lettuce", "jalapeno", "kale", "leek", "mushroom", "nectarine", "olive", "parsnip", "quince", "radish", "shallot", "tangerine", "ugli fruit", "vanilla bean", "watercress", "yam",
               "zucchini", "apple", "banana", "cherry", "date", "elderberry", "fig", "grape", "honeydew", "kiwi", "lemon", "mango", "nectarine", "orange", "pear", "quince", "raspberry", "strawberry", "tangerine", "watermelon", "apricot", "blackberry", "cantaloupe", "dragonfruit", "grapefruit", "jackfruit", "kiwifruit", "lime", "mulberry", "olive", "papaya", "quince", "rhubarb", "starfruit", "tamarind", "ugli fruit", "vanilla bean", "yuzu", "zucchini", "almond", "cashew", "date", "eggplant", "fig", "garlic", "hazelnut", "iceberg lettuce", "jalapeno", "kale", "leek", "mushroom", "nectarine", "olive", "parsnip", "quince", "radish", "shallot", "tangerine", "ugli fruit", "vanilla bean", "watercress", "yam", "zucchini"]

createdTimes: list = ["2024-04-06 10:05:00", "2024-04-06 10:40:00", "2024-04-06 10:20:00",
                      "2024-04-06 10:48:00", "2024-04-06 10:14:00", "2024-04-06 10:16:00"]

# callbackul pt delivery report


def delivery_report(err, msg):
    if err is not None:
        print(f'Message delivery failed: {err}')
    else:
        print(f'Message delivered to {msg.topic()} [{msg.partition()}]')

# json serializer


def json_serializer(data):
    return json.dumps(data).encode('utf-8')


p = Producer({'bootstrap.servers': KAFKA_BOOTSTRAP_SERVER})


def produce():
    try:
        while True:
            event = {
                "word": random.choice(words),
                "createdAt": random.choice(createdTimes)
            }
            p.produce(KAFKA_TOPIC,
                      value=json_serializer(event),
                      key=uuid.uuid4().hex,
                      callback=delivery_report,
                      headers=[("content-type", "application/json")]
                      )
            # asteapta ca mesajul sa fie trimis, in caz contrar se blocheaza
            p.poll(0)

            time.sleep(random.randint(1, 5))
    finally:
        # asteapta ca mesajele sa fie trimise inainte de a inchide producerul
        p.flush()


produce()
```
